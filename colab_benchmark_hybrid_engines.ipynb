# Hybrid LNN Engine-Level Benchmark Evaluation

This notebook evaluates the current `hybrid_full_xtb` checkpoint on the external benchmark sets and scores each site engine separately:
- `final` fused output
- `base_lnn` backbone site logits
- `lnn_vote`
- `wave_vote`
- `analogical_vote`

It reports metrics for:
- all molecules
- wave-valid subset
- wave-missing subset
