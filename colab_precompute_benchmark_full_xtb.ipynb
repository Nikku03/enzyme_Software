# Precompute full-xTB for External Benchmarks A/B/C

This notebook precomputes the `full_xtb` cache used by the hybrid LNN wave path for:
- External Benchmark A: row-level single-CYP benchmark
- External Benchmark B: unique molecules only
- External Benchmark C: high-confidence subset only

It writes cached payloads into Google Drive and saves a JSON summary of which molecules succeeded.

```python
# Cell 1: clone/pull repo and install package
import os, subprocess, sys

REPO = 'https://github.com/Nikku03/enzyme_Software.git'
REPO_DIR = '/content/enzyme_Software'

if not os.path.exists(REPO_DIR):
    subprocess.check_call(['git', 'clone', REPO, REPO_DIR])
else:
    subprocess.check_call(['git', '-C', REPO_DIR, 'pull', '--ff-only'])

subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-q', '-e', REPO_DIR])
print('Repo ready:', REPO_DIR)
```

```python
# Cell 2: mount Google Drive for persistent cache / summaries
import os
from google.colab import drive

drive.mount('/content/drive')
os.makedirs('/content/drive/MyDrive/enzyme_hybrid_lnn/cache/full_xtb', exist_ok=True)
os.makedirs('/content/drive/MyDrive/enzyme_hybrid_lnn/artifacts/hybrid_full_xtb', exist_ok=True)
print('Drive ready.')
```

```python
# Cell 3: configure benchmark datasets / output paths
import os

os.environ['HYBRID_COLAB_BENCHMARK_XTB_CACHE_DIR'] = '/content/drive/MyDrive/enzyme_hybrid_lnn/cache/full_xtb'
os.environ['HYBRID_COLAB_BENCHMARK_XTB_DATASETS'] = ','.join([
    'data/prepared_training/main8_benchmark_a_row_level_singlecyp.json',
    'data/prepared_training/main8_benchmark_b_unique_molecules.json',
    'data/prepared_training/main8_benchmark_c_high_confidence.json',
])
os.environ['HYBRID_COLAB_BENCHMARK_XTB_SUMMARY_JSON'] = '/content/drive/MyDrive/enzyme_hybrid_lnn/artifacts/hybrid_full_xtb/benchmark_full_xtb_summary.json'

print('HYBRID_COLAB_BENCHMARK_XTB_CACHE_DIR=', os.environ['HYBRID_COLAB_BENCHMARK_XTB_CACHE_DIR'])
print('HYBRID_COLAB_BENCHMARK_XTB_DATASETS=', os.environ['HYBRID_COLAB_BENCHMARK_XTB_DATASETS'])
print('HYBRID_COLAB_BENCHMARK_XTB_SUMMARY_JSON=', os.environ['HYBRID_COLAB_BENCHMARK_XTB_SUMMARY_JSON'])
```

```python
# Cell 4: run benchmark full-xTB precompute
%cd /content/enzyme_Software
exec(open('/content/enzyme_Software/scripts/colab_precompute_benchmark_full_xtb.py').read(), {'__name__': '__main__'})
```

```python
# Cell 5: inspect the saved summary JSON
import json

summary_path = os.environ['HYBRID_COLAB_BENCHMARK_XTB_SUMMARY_JSON']
with open(summary_path) as f:
    summary = json.load(f)
print(json.dumps(summary, indent=2))
```
