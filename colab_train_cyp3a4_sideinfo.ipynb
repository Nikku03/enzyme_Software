# CYP3A4 Sideinfo Training

CYP3A4-only training with the LNN as the sole predictor and NEXUS wave/analogical used only as side-information features.

```python
from google.colab import drive
drive.mount('/content/drive')

import os
import subprocess
from pathlib import Path

REPO_DIR = Path('/content/enzyme_Software')
REPO_URL = 'https://github.com/Nikku03/enzyme_Software.git'

if not REPO_DIR.exists():
    subprocess.check_call(['git', 'clone', REPO_URL, str(REPO_DIR)])
else:
    subprocess.check_call(['git', 'fetch', 'origin'], cwd=str(REPO_DIR))
    subprocess.check_call(['git', 'checkout', 'main'], cwd=str(REPO_DIR))
subprocess.check_call(['git', 'pull', 'origin', 'main'], cwd=str(REPO_DIR))
subprocess.check_call(['python', '-m', 'pip', 'install', '-r', 'requirements.txt'], cwd=str(REPO_DIR))
os.chdir(str(REPO_DIR))
print('repo=', REPO_DIR)
```
