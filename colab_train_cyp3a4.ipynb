# NEXUS CYP3A4 Training on Colab

Run this notebook top-to-bottom.

What it does:
- trains NEXUS only on `3A4.sdf`
- keeps the broader analogical memory bank
- masks exact-query retrieval so the memory bank cannot return the same molecule
- uses the dedicated wrapper script `scripts/colab_train_cyp3a4.py`


```python
# Cell 1: clone/pull repo and install package
import os, subprocess, importlib

REPO = 'https://github.com/Nikku03/enzyme_Software.git'
REPO_DIR = '/content/enzyme_Software'

if not os.path.exists(REPO_DIR):
    print('Cloning repo...')
    subprocess.run(f'git clone {REPO} {REPO_DIR}', shell=True, check=True)
else:
    print('Repo exists, pulling latest...')
    subprocess.run(f'git -C {REPO_DIR} pull', shell=True, check=True)

os.chdir(REPO_DIR)
print('Working dir:', os.getcwd())

subprocess.run('pip install rdkit -q', shell=True, check=True)
subprocess.run('pip install -e . -q', shell=True, check=True)
print('Install complete.')
```

```python
# Cell 3: choose preset / optional overrides
import os

# Presets:
#   fast          - quick debug (32 molecules, 3 epochs)
#   balanced      - default iteration-speed / fidelity balance (64 molecules, 4 epochs)
#   full_3a4      - all 457 molecules, 5 epochs, trimmed physics
#   full_3a4_a100 - all 457 molecules, 25 epochs, full physics for A100/H100 80GB
#   rtx6k_2h      - all 457 molecules, 20 epochs, ~2 hours on RTX 6000 Ada (95 GB)
#                   uses: torch.compile + TF32 + physics curriculum + bg data loading
#                   (single-GPU only — if you have 8×A100, target one node)
os.environ['NEXUS_COLAB_RUN_PRESET'] = 'rtx6k_2h'

# Optional per-key overrides (take priority over the preset above):
# os.environ['NEXUS_COLAB_EPOCHS'] = '20'
# os.environ['NEXUS_COLAB_MAX_SAMPLES'] = '0'        # 0 = all molecules
# os.environ['NEXUS_COLAB_ALLOW_COMPILE'] = '1'      # torch.compile SIREN (safe on RTX/A100)
# os.environ['NEXUS_COLAB_NUM_WORKERS'] = '2'        # background data loading
# os.environ['NEXUS_COLAB_CURRICULUM'] = '1'         # progressive physics ramp
# os.environ['NEXUS_COLAB_DYNAMICS_STEPS'] = '1'
# os.environ['NEXUS_COLAB_NAV_OPT_STEPS'] = '1'
# os.environ['NEXUS_COLAB_NAV_CANDIDATES'] = '2'
# os.environ['NEXUS_COLAB_GPU_PROFILE'] = 'ultra_vram'

for key in [
    'NEXUS_COLAB_RUN_PRESET',
    'NEXUS_COLAB_GPU_PROFILE',
    'NEXUS_COLAB_TARGET_ISOFORM',
    'NEXUS_COLAB_MAX_SAMPLES',
    'NEXUS_COLAB_EPOCHS',
    'NEXUS_COLAB_DYNAMICS_STEPS',
    'NEXUS_COLAB_NAV_OPT_STEPS',
    'NEXUS_COLAB_NAV_CANDIDATES',
    'NEXUS_COLAB_ALLOW_COMPILE',
    'NEXUS_COLAB_NUM_WORKERS',
    'NEXUS_COLAB_CURRICULUM',
]:
    if key in os.environ:
        print(f'{key}={os.environ[key]}')
```
